```python
import matplotlib
import astropy.units as u
from astropy.coordinates import SkyCoord
from toast import QueryWidget # the on-sky area stacking tool
```

```python
### Main Widget Usage ###

# Usage order: Lasso Select --> Confirm Selection --> Extract Spectra (--> Reset)
# You can use a SkyCoord object or a Detectid. If both are given, it will use the Detectid. 

try:
    qw.close() # this is *supposed to* close/clear any remaining axes...doesn't always. fix later.
except:
    pass

qw = QueryWidget(#detectid = None,
                 coords = SkyCoord(210.804* u.deg, 54.349* u.deg, frame="icrs"), #M101 coordinates
                 cutout_size=10*u.arcsec, 
                 spec_mode='psf', # 'fibers', 'psf' #psf naturally takes longer
                # wave_range=(5230, 5400), 
                # flux_range=(-0.2, 0.4),
                 stat='weighted_biweight', # 'mean', 'median', 'biweight', 'weighted_biweight'
                 )

# image loading is not always immediate. matplotlib has to redraw the figure for every pan/lasso draw. please be patient. If the fiber locs
# do not appear, they should upon clicking "lasso select"

# to select different fibers, simply redraw the lasso. Reset button will pan back to original coords. 

# NOTE: if you use the up/down arrows on RA/Dec, it will truncate to the set step (i.e. be less exact). 
# Be careful when recording coordinates.
```
