```python
import matplotlib
import astropy.units as u
from astropy.coordinates import SkyCoord
from toast import QueryWidget # the on-sky area stacking tool - extended region (i.e. for all of HETDEX)
```

```python

### Main Widget Usage ###

# Usage order: Lasso Select --> Confirm Selection --> Extract Spectra (--> Reset)
# You can use a SkyCoord object or a Detectid. If both are given, it will use the Detectid. 


try:
    qw.close() # this is *supposed to* close/clear any remaining axes...doesn't always. fix later.
except:
    pass

qw = QueryWidget(coords = SkyCoord(210.8* u.deg, 54.35* u.deg, frame="icrs"),
                 cutout_size=20*u.arcsec, 
                 spec_mode='fibers', # 'fibers', 'psf'
                # wave_range=(5230, 5400), 
                # flux_range=(-0.2, 0.4),
                 stat='weighted_biweight', # 'mean', 'median', 'biweight', 'weighted_biweight'
                 )

# Matplotlib is a pain when redrawing axes (necessary for laso selection). 
# Don't click until black goes away. If the image isn't displaying properly, try the following:
# - Click reset, try drawing a lasso, click reset again.
# - Run qw.close() in next cell. Then re-run QueryWidget call
# - Shut down kernel, clear all outputs and restart

# NOTE: if you use the up/down arrows on RA/Dec, it will truncate to the set step (i.e. be less exact). 
# Be careful when recording coordinates.
```
